```python
# 07 kwargs

def suma(a, b):
  return a + b
print(suma(4, 5))  
```

```python
print(suma(4, 5, 6))  
```

```python
def suma(*args):
  sum = 0
  for x in args:
    sum +=x
  return sum  
print(suma(4, 5, 6))    
```

```python
print(suma(cafe = 10, jugo = 20, tostado = 25))
```

```python
def suma(**kwargs):
  sum = 0
  for x in kwargs.values():
    sum +=x
  return sum  
print(suma(cafe = 10, jugo = 20, tostado = 25))
```

```python
# 09 - Manejo de excepciones

def dividir(a, b):
  return a / b
```

```python
print(dividir(40, 4))
```

```python
print(dividir(40, 0))
```

```python
try:
   respuesta = (dividir(40, 0))
except:
  print("Algo sali√≥ mal")   
```

```python
try:
   respuesta = (dividir(40, 0))
except Exception as e:
  print("Algo sali√≥ mal:", e)   
  print(print(e.__class__))
```

```python
try:
   respuesta = (dividir(40, 0))
except ZeroDivisionError as e:
  print(e, "No se puede dividir por cero") 
```

```python
try:
   respuesta = (dividir(40, 0))
except ZeroDivisionError as e:
  print(e, "No se puede dividir por cero") 
except Exception as e:
  print("Algo sali√≥ mal:", e)     
```
