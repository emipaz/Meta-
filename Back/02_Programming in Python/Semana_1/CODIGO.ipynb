```python
# 15 - Control de flujo

cuenta_total = 95 # lo cambio
descuento1 = 10

if cuenta_total > 100:
  print("Le cuenta es mayor que 100")
  cuenta_total -= descuento1 
else:  
  print("Le cuenta es menor que 100")

print("La cuenta es de: " + str(cuenta_total))


```

```python
cuenta_total = 210 # lo cambio
descuento1 = 10
descuento2 = 20
if cuenta_total > 100 and cuenta_total < 200:
  print("La cuenta es mayor que 100")
  cuenta_total -= descuento1 
elif cuenta_total > 200:
  print("La cuenta es mayor que 200")
  cuenta_total -= descuento2
else:  
  print("Le cuenta es menor que 100")

print("La cuenta es de: " + str(cuenta_total))
```

```python
# 16 - Sentencia switch

http_status = 200

if http_status == 200 or http_status == 201:
  print("¡Todo bien!")
elif http_status == 400:
  print("Bad request")
elif http_status == 404:
  print("Not found")
elif http_status == 500 or http_status == 501:
  print("Server error")
else:
  print("Unknown")
```

```python
http_status = 200

match http_status:
  case 200 | 201:
      print("¡Todo bien!")
  case 400:
      print("Bad request")
  case 404:
      print("Not found")
  case 500 | 501:
      print("Server error")
  case _:
      print("Unknown")
```

```python
# 17 - Loops
favorites = ["Creme Brulee", "Apple pie", "Churros", "Tiramisu", "Chocolate cake"]

for i in range(10):
  print("Looping...", i)
```

```python
for item in favorites:
  print("Me gusta este postre", item)
```

```python
contador = 0
while contador < len(favorites):
  print("Me gusta este postre", favorites[contador])
  contador +=1
```

```python
for idx, item in enumerate(favorites):
  print(idx, item)
```
